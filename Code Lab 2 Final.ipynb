```python
#Andrew Lonie Lab 2
#Geoprocessing Tool
#Approx Code Time: 18 Hours 

import os
import sys
sys.path.append('C:\\Program Files (x86)\\ArcGIS\\Desktop10.3\\bin')
sys.path.append('C:\\Program Files (x86)\\ArcGIS\\Desktop10.3\\arcpy')
sys.path.append('C:\\Program Files (x86)\\ArcGIS\\Desktop10.3\\ArcToolbox\\Scripts')
import arcpy
from arcpy import env
env.overwriteOutput = True               
                    
env.workspace = r"C:\Geog458\Lab2\geoprocessing"
                    
#This part was pretty quick
#I just checked the input shapefile along with checking
#the reclass table.
def reclassTool(mytable, infield, outfield, reclasstable, notfoundvalue):
    
    if arcpy.Exists(mytable) == True:
        
        if arcpy.Exists(reclasstable) == True:

           
            arcpy.AddField_management(mytable, outfield, "TEXT", "10", ")
       
            
            cursor1 = arcpy.da.UpdateCursor(mytable, [infield, outfield])
#made my list here of values with the reclass table
#Little emotional support from Jeremy here with this.
#I was debating on whether or not to use a search or insert cursor here.
            values = []
            
            for row1 in cursor1:
                
                 
                cursor2 = arcpy.da.SearchCursor(reclasstable, ["lower", "upper", "totalvalue"])
        
               
                for row2 in cursor2:

                    if row2[2] not in values:
                        values.append(row2[2])

                del row2
                del cursor2
       
                
                cursor2 = arcpy.da.SearchCursor(reclasstable, ["lower", "upper", "totalvalue"])
                for row2 in cursor2:
#here we determine all of the value classification we need.
#We are looking for an infield value that is greater/equal
#This part of the project provides us with the numbers and 
#classification we need to determine the reclass table. 
                    if row1[0] >= row2[0] and row1[0] <= row2[1]:
                        row1[1] = row2[2]
                    else:
                        pass

                            
                if row1[1] not in values:

                    row1[1] = notfoundvalue
                else:
                    pass

#Always remembering to delete cursors and rows @ end of code.            
            del row1
            del row2
            del cursor1
            del cursor2
```
